```python
import numpy as np
import pandas as pd
from influxdb import InfluxDBClient
```

```python
# Connect to InfluxDB and fetch data
client = InfluxDBClient(host='localhost', port=8086)
client.switch_database('ISS')
```

```python
# Query to the Database for one measurement

#query1 = 'SELECT * FROM "DE1Thing_HMD8310"'
#query1 = 'SELECT * FROM "DE1Thing_HMD8310" WHERE time >= \'2022-08-29T23:28:00Z\' AND time < \'2023-01-26T08:00:00Z\''
query1 = 'SELECT * FROM "EP_Battery.Thing_HMD8310"'
results1 = client.query(query1)
EP_Battery_Thing_HMD8310 = pd.DataFrame.from_records(results1.get_points())

print(EP_Battery_Thing_HMD8310.shape)
EP_Battery_Thing_HMD8310.head(5)
```

```python
EP_Battery_Thing_HMD8310.to_csv('EP_Battery_Thing_HMD8310.csv', index=False)
```
